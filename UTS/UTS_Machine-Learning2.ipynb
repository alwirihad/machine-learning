# Import Library 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Data

```python
fakultas = ["Bisnis", "D3 Perhotelan", "ICT", "Ilmu Komunikasi", "Seni dan Desain"]
jumlah_mahasiswa = [260, 28, 284, 465, 735]
```

# Membuat DataFrame

```python
data = pd.DataFrame({
    'Fakultas': fakultas,
    'Jumlah Mahasiswa': jumlah_mahasiswa
})
```

# Menampilkan data

```python
print(data)
```

# Membuat dan Menampilkan Plot

```python
warna = ["pink", "brown", "green", "blue", "purple"]
plt.figure(figsize=(10, 6))
plt.bar(data['Fakultas'], data['Jumlah Mahasiswa'], color=warna)
plt.title('Jumlah Mahasiswa per Fakultas')
plt.xlabel('Fakultas')
plt.ylabel('Jumlah Mahasiswa')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
