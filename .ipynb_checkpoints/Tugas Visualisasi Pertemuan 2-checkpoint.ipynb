# Soal 2

Untuk Masalah ini, saya memilih menggunakan pie chart untuk memvisualisasikan presentase peluang mengeluarkan permen snickers dari kantong jika kita melakukan pengambilan secara acak.

```python
import matplotlib.pyplot as plt

# Data
candy_names = ['Kit Kat', 'Snickers', 'Milky Way', 'Toblerone', 'Twix']
candy_counts = [52, 39, 78, 13, 78]

# Persentase peluang Snickers
snickers_index = candy_names.index('Snickers')
snickers_probability = candy_counts[snickers_index] / sum(candy_counts) * 100

# Labels
labels = candy_names
sizes = candy_counts

# Colors
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99', '#c2c2f0']
explode = [0, 0.1, 0, 0, 0]

# Creating the pie chart
plt.figure(figsize=(8, 6))
plt.pie(
    sizes, 
    labels=labels, 
    colors=colors, 
    autopct='%1.1f%%', 
    startangle=140,
    explode=explode,
    shadow=True,
)

# Title
plt.title('Persentase Peluang Mengeluarkan Permen Snickers')

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')

# Display the plot
plt.show()

```

Terdapat peluang 15% dalam menentukan untuk mengeluarkan snickers dari dalam tas

# Soal 3

Untuk masalah ini, saya memilih menggunakan bar chart. Karena bar chart memungkinkan kita untuk dengan mudah membandingkan nilai numerik antara beberapa kategori, yang sesuai dengan kasus ini di mana kita membandingkan penjualan makanan penutup.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Data
dessert_sales = {
  'Lava Cake': 14,
  'Mousse': 5,
  'Chocolate Cake': 12,
  'Ice Cream': 19,
  'Truffles': 6,
  'Brownie': 8,
  'Chocolate Chip Cookie': 12,
  'Chocolate Pudding': 9,
  'Souffle': 10,
  'Chocolate Cheesecake': 17,
  'Chocolate Chips': 2,
  'Fudge': 9,
  'Mochi': 13,
}

# Sorting desserts by sales (ascending order)
sorted_desserts = sorted(dessert_sales.items(), key=lambda x: x[1])

# Extracting dessert names and sales
desserts = [dessert[0] for dessert in sorted_desserts]
sales = [dessert[1] for dessert in sorted_desserts]

# Creating the DataFrame
df = pd.DataFrame({
    'desserts': desserts,
    'sales': sales,
})
df.sort_values(by='sales', inplace=True)

# Creating the bar chart with colors for the lowest 3 desserts
plt.figure(figsize=(10, 6))
bars = plt.bar(df['desserts'], df['sales'], color='skyblue')

# Coloring the lowest 3 bars in red
for i in range(3):
    bars[i].set_color('red')

# Adding labels and title
plt.xticks(rotation=90)
plt.ylabel('Penjualan')
plt.xlabel('Makanan Penutup')
plt.title('Popularitas Relatif Makanan Penutup')

# Displaying the plot
plt.show()
```

Makanan penutup yang harus disingkirkan adalah
- Chocolate Chips
- Mousse
- Truffles

Dikarenakan makanan tersebut tidak banyak yang ingin memesan untuk penutupnya

# Soal 4

Untuk masalah ini, saya menggunakan heatmap. Dikarenakan heatmap cocok digunakan untuk menampilkan data dalam format matriks di mana intensitas warna menunjukkan nilai data.

```python
import matplotlib.pyplot as plt

# Data
days = ['Senin', 'Selasa', 'Rabu', 'Kamis', 'Jumat', 'Sabtu', 'Minggu']
hours = list(range(24))
cpu_usage = [
  [2, 2, 4, 2, 4, 1, 1, 4, 4, 12, 22, 23, 45, 9, 33, 56, 23, 40, 21, 6, 6, 2, 2, 3], # Monday
  [1, 2, 3, 2, 3, 2, 3, 2, 7, 22, 45, 44, 33, 9, 23, 19, 33, 56, 12, 2, 3, 1, 2, 2], # Tuesday
  [2, 3, 1, 2, 4, 4, 2, 2, 1, 2,  5, 31, 54, 7, 6, 34, 68, 34, 49, 6, 6, 2, 2, 3], # Wednesday
  [1, 2, 3, 2, 4, 1, 2, 4, 1, 17, 24, 18, 41, 3, 44, 42, 12, 36, 41, 2, 2, 4, 2, 4], # Thursday
  [4, 1, 2, 2, 3, 2, 5, 1, 2, 12, 33, 27, 43, 8, 38, 53, 29, 45, 39, 3, 1, 1, 3, 4], # Friday
  [2, 3, 1, 2, 2, 5, 2, 8, 4, 2, 3, 1, 5, 1, 2, 3, 2, 6, 1, 2, 2, 1, 4, 3], # Saturday
  [1, 2, 3, 1, 1, 3, 4, 2, 3, 1, 2, 2, 5, 3, 2, 1, 4, 2, 45, 26, 33, 2, 2, 1], # Sunday
]

# Creating the heat map
plt.figure(figsize=(10, 6))
plt.imshow(cpu_usage, cmap='hot', interpolation='nearest')

# Adding labels
plt.xticks(range(len(hours)), hours)
plt.yticks(range(len(days)), days)
plt.xlabel('Jam (0 = Tengah Malam)')
plt.ylabel('Hari')

# Adding color bar
plt.colorbar(label='Penggunaan CPU')

# Title
plt.title('Penggunaan CPU per Jam selama Seminggu')

# Displaying the plot
plt.show()
```
